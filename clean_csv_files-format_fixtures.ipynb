```python
# Equipos confirmados para clasificacion
teams_2026_confirmed = [
    # Anfitriones (CONCACAF)
    "United States", "Mexico", "Canada",
    # Sudamérica (CONMEBOL)
    "Argentina", "Brazil", "Ecuador", "Uruguay", "Colombia", "Paraguay",
    # Asia (AFC)
    "Japan", "Iran", "Uzbekistan", "South Korea", "Jordan", "Australia", "Saudi Arabia", "Qatar",
    # África (CAF)
    "Algeria", "Egypt", "Ghana", "Cape Verde", "Ivory Coast", "Morocco", "Senegal", "South Africa", "Tunisia",
    # Europa (UEFA)
    "England", "France", "Croatia",
    # Oceanía (OFC)
    "New Zealand"
]
# Posibles 18 clasificados faltantes
projected_remaining_teams = [
    # UEFA
    "Germany", "Spain", "Portugal", "Netherlands", "Belgium",
    "Italy", "Switzerland", "Denmark", "Serbia",
    "Poland", "Scotland", "Turkey", "Austria",
    # CONMEBOL
    "Chile", "Peru",
    # CAF
    "Nigeria", "Mali",
    # CONCACAF
    "Panama"
]

# Grupos desde la A a la L(48 equipos participantes) 
grupos_2026 = {} 
for letra in "ABCDEFGHIJKL": 
    grupos_2026[f"Grupo {letra}"] = [ 
        {"Pos": 1, "Team": f"{letra}1", "Pld": 0, "W": 0, "D": 0, "L": 0, "GF": 0, "GA": 0, "GD": 0, "Pts": 0}, 
        {"Pos": 2, "Team": f"{letra}2", "Pld": 0, "W": 0, "D": 0, "L": 0, "GF": 0, "GA": 0, "GD": 0, "Pts": 0}, 
        {"Pos": 3, "Team": f"{letra}3", "Pld": 0, "W": 0, "D": 0, "L": 0, "GF": 0, "GA": 0, "GD": 0, "Pts": 0}, 
        {"Pos": 4, "Team": f"{letra}4", "Pld": 0, "W": 0, "D": 0, "L": 0, "GF": 0, "GA": 0, "GD": 0, "Pts": 0} 
    ]

# Lista con los equipos finales para el Mundial 2026
final_teams_2026 = teams_2026_confirmed + projected_remaining_teams

# CREAR LOS BOMBOS FIFA (POT 1–4)
pot1 = [
    "Mexico", "Canada", "United States",  # Hosts
    "Argentina", "Brazil", "France", "England",
    "Portugal", "Spain", "Germany",
    "Netherlands", "Croatia"
]

pot2 = [
    "Uruguay", "Colombia", "Japan", "Morocco",
    "Italy", "Switzerland", "Denmark", "Belgium",
    "Senegal", "South Korea", "Australia", "Chile"
]

pot3 = [
    "Ecuador", "Paraguay", "Iran", "Uzbekistan",
    "Saudi Arabia", "Qatar", "Egypt", "Algeria",
    "Ghana", "Ivory Coast", "South Africa", "Tunisia"
]

pot4 = [
    "Cape Verde", "New Zealand", "Serbia", "Poland",
    "Scotland", "Turkey", "Austria", "Peru", 
    "Nigeria", "Mali", "Panama", "Jordan"
]

# Paises que estan de Anfitriones
grupos_2026["Grupo A"][0]["Team"] = "Mexico"
grupos_2026["Grupo B"][0]["Team"] = "Canada"
grupos_2026["Grupo D"][0]["Team"] = "United States"

# Asignar de manera aleatoria los equipos a los grupos
random.seed(69)

remaining_pot1 = [t for t in pot1 if t not in ["Mexico", "Canada", "United States"]]

i = 0
for letra in "ACDEFGHIJKL":  # excluyendo B (ya tiene)
    # La posición 0 (Pos 1) ya está asignada a host, así que saltamos
    if letra != "A" and letra != "B" and letra != "D":
        grupos_2026[f"Grupo {letra}"][0]["Team"] = remaining_pot1[i]
        i += 1

# Funcion para asignar los Bombos en las posciones (2, 3, 4)
def assign_pot_to_position(pot, position):
    random.shuffle(pot)
    idx = 0
    for letra in "ABCDEFGHIJKL":
        # position = 2,3,4 → índice = position - 1
        grupos_2026[f"Grupo {letra}"][position-1]["Team"] = pot[idx]
        idx += 1
        
assign_pot_to_position(pot2, 2)
assign_pot_to_position(pot3, 3)
assign_pot_to_position(pot4, 4)
```

```python
# cargamos los matches ordenados
df_fixture_matches = df_fixture['score'].unique()

# Generamos los partidos por grupo
def generar_fixture_grupo(group_name, teams_list):
    t1 = teams_list[0]["Team"]
    t2 = teams_list[1]["Team"]
    t3 = teams_list[2]["Team"]
    t4 = teams_list[3]["Team"]

    partidos = [
        (t1, t2),
        (t3, t4),
        (t1, t3),
        (t2, t4),
        (t1, t4),
        (t2, t3),
    ]

    df = pd.DataFrame(partidos, columns=["home", "away"])
    df["group"] = group_name
    return df


# Unimos todos los grupos
fixture_total = pd.concat(
    [generar_fixture_grupo(g, grupos_2026[g]) for g in grupos_2026],
    ignore_index=True
)

#Ahora asignamos los scores del fixture real a los 6 matches por grupo:
matches_por_grupo = [
    df_fixture_matches[i:i+6] 
    for i in range(0, 72, 6)
]

# Asignamos los Match X a los grupos
fixture_total["score"] = None
i = 0
for idx_grupo, group_name in enumerate(grupos_2026):
    fixture_total.loc[i:i+5, "score"] = matches_por_grupo[idx_grupo]
    i += 6

# Crear 'year' column and drop 'group' column
fixture_total["year"] = 2026
fixture_total = fixture_total.drop(['group'], axis=1)

# Reordenar columnas
fixture_total = fixture_total[['home', 'score', 'away', 'year']]

# Verificar
fixture_total.head()
```
